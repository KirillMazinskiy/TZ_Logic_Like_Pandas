```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Lenovo\Downloads\data.csv', encoding = 'utf-8', sep = ',')
```

```python
df.clientID.nunique()
```

```python
def get_client_type(row):
    goalsID = row['goalsID']
    if goalsID == 25230759.:
        return 'Новый'

    return 'Старый'
```

```python
def get_tarif(row):
    goalsID = row['goalsID']
    if goalsID ==  40343059.:
        return 1

    return 0
```

```python
def get_active_client(row):
    if "process" in row['new_url']:
        return "1"
    return "0"
    
```

```python
def get_unic_task(row):
    if "result" in row['new_url']:
        return row['new_url'][-2]



```

```python
def get_target(row):
    if "'targetLessons'" in row['paramsKeys']:
        return row['paramsValues'][row['paramsKeys'].index("'targetLessons'")]

    return "'-1'"
```

```python
df['paramsValues'] = df['paramsValues'].map(lambda x: x.strip('][').split(','))
```

```python
df['paramsKeys'] = df['paramsKeys'].map(lambda x: x.strip('][').split(','))
```

```python
df['new_url'] = df['URL'].map(lambda x: x.strip('][').split('/'))
```

```python
df['true_target'] = df.apply(get_target, axis=1)
```

```python
df['numberOfUnicTask'] = df.apply(get_unic_task, axis=1)
```

```python
df['client_status'] = df.apply(get_active_client, axis=1)
```

```python
df['client_type'] = df.apply(get_client_type, axis=1)
```

```python
df['tarif_wachted'] = df.apply(get_tarif, axis=1)
```

```python
ser  = df\
.groupby(['clientID'], as_index =False)\
.agg({'true_target':'max', 'client_type':'min','client_status':'max', 'tarif_wachted':'max'})
```

```python
ser = ser.drop(ser[ser['true_target'] == "'0'"].index)
```

```python
ser
```

```python
file = ser\
.rename(columns = {'client_type':'Тип пользователя', 'clientID':'Количество клиентов', 'true_target':'Возрастная когорта'})\
.groupby(['Возрастная когорта', 'Тип пользователя'])\
.agg({'Количество клиентов':'count', 'tarif_wachted':'sum'})
```

```python
file
```

```python
file['Конверсия в просмотр тарифов'] = file['tarif_wachted']/file['Количество клиентов'] * 100
```

```python
file
```

```python
count_task = df\
.groupby(['clientID'], as_index = False)\
.agg({'true_target':'max', 'numberOfUnicTask':'nunique', 'client_type':'min','client_status':'max'})\
.rename(columns={'numberOfUnicTask':'sum_completed_unic_task'})
```

```python
count_task =count_task.drop(count_task[count_task['true_target'] == "'0'"].index)
```

```python
count_task
```

```python
srednee_chislo = count_task\
.query('client_status == "1"')\
.groupby(['true_target', 'client_type'])\
.agg({'clientID':'count', 'sum_completed_unic_task':'sum'})
```

```python
srednee_chislo
```

```python
file['Среднее число решенных задач'] = srednee_chislo['sum_completed_unic_task'] / srednee_chislo['clientID']
```

```python
file
```

```python
final_file = file\
.drop(columns = {'tarif_wachted'})
```
